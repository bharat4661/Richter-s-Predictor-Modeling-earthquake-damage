```python
from sklearn.metrics import f1_score
pipe = make_pipeline(StandardScaler(), RandomForestClassifier(random_state=45))

param_grid = {'randomforestclassifier__n_estimators': [25],
              'randomforestclassifier__min_samples_leaf': [5],
              'randomforestclassifier__n_jobs': [-1]}
gs = GridSearchCV(pipe, param_grid, cv=3)
training_start = time.perf_counter()
gs.fit(x,y.values.ravel())
prediction_start = time.perf_counter()
in_sample_preds = gs.predict(x)
prediction_end = time.perf_counter()

acc_rfc = f1_score(y,in_sample_preds,average='micro')
rfc_train_time = training_end-training_start
rfc_prediction_time = prediction_end-prediction_start
print("Scikit-Learn's Random Forest Classifier's prediction accuracy is: %3.2f" % (acc_rfc))
print("Time consumed for training: %4.3f seconds" % (rfc_train_time))
print("Time consumed for prediction: %6.5f seconds" % (rfc_prediction_time))
print(gs.best_params_)
confusion_matrix(y,in_sample_preds)
```
